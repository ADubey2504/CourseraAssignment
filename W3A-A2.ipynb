## Import Beautiful Soup

```python
from urllib.request import urlopen
import requests
from bs4 import BeautifulSoup

print('Done')
```

```python
url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
```

### Map URL and interpret HTML Table

```python
soup = BeautifulSoup(url,'lxml')
type(soup)
#print(soup.prettify())
```

## Identify Postal Code Table

```python
My_table = soup.find('table',{'class':'wikitable sortable'})
#My_table
```

```python
import csv
```

## Fetch Content from HTML Table into Rows and Columns

```python
list_of_rows= []   
rows  = My_table.find_all('tr')

for row in rows:
    list_of_cells = []
    for cell in row.find_all('td'):
        list_of_cells.append(cell.text)
    list_of_rows.append(list_of_cells)
list_of_rows
```

```python
outfile=open("./postcodes.csv","w")
writer =csv.writer(outfile)
writer.writerow(["Postal Code","Borough","Neighbourhood"])
writer.writerows(list_of_rows)
```

### Generate CSV file and migrate the HTML List into CSV file

```python
#import necessary modules

with open('./postcodes.csv','rt')as f:
  data = csv.reader(f)
  for row in data:
        print(row)
```

### Read CSV File into Data frame

```python
import pandas as pd
result=pandas.read_csv('./postcodes.csv')
df = pd.DataFrame(result)
df
```

## Clean the table by removing EoL

```python
df['Neighbourhood'].replace(r'\s+|\\n', ' ', regex=True, inplace=True)
print(df)
```

# Remove Boroughs from data frame where value is Not assigned

```python
df1= df[df['Borough']!= 'Not assigned']
df1
```

## 77 rows were not considered where Borough = Not Assigned

```python
df1.sort_values(by=['Postal Code'])
```

## CHeck rows where Neighbourhood values is Not Assigned

```python
print("Check Not Assigned values in Neighbourhood column:\n") 
df2 = df1['Neighbourhood'].str.contains('Not assigned') 
df2
```

###Results shows there are no rows where Neighbourhood value is not assigned 

## Now merge neighbourhood with same Post codes and Borough names into single row

```python
df2 = (df1.groupby(['Postal Code', 'Borough'])['Neighbourhood'].apply(list))
df2 
```
